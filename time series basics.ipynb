# custom holiday day indian calendar 26 jan

```python
import pandas as pd
df = pd.read_csv("TATAPOWER.NS3.csv")
df
```

```python
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
from pandas.tseries.offsets import CustomBusinessDay

class indianHolidayCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('RepublicDay', month=1, day=26)
    ]
    
mnc = CustomBusinessDay(calendar=indianHolidayCalendar())
mnc
```

```python
rng = pd.date_range(start="1/1/2022", end="31/1/2022", freq=mnc)
rng
```

```python
df.set_index(rng, inplace = True)
df
```

# or you can do this

```python
b = CustomBusinessDay(holidays=["2022-01-26"])
rng = pd.date_range(start="1/1/2022", end="31/1/2022", freq=b)
rng

                      
```

```python
df.set_index(rng, inplace = True)
df
```

## shifting and lagging

```python
import pandas as pd
df = pd.read_excel("demo.xlsx", parse_dates=['Date'], index_col='Date')
df
```

```python
df.shift(1)
```

```python
df.shift(-1)
```

```python
df['Prev Day price'] = df['Price'].shift(1)
df
```

```python
df['1 Day change']= df['Price'] - df['Prev Day price']
df
```

```python
df['5 day % return'] =(df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

```python
df = df[['Price']]
df
```
